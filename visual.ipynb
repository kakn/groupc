```python
import matplotlib.pyplot as plt
import numpy as np

def read_csv(filename):
    infile = open(filename, 'r')
    lines = infile.readlines()
    lines = [line.split(',') for line in lines]
    return lines

male = 0
female = 0

lines = read_csv('blogtext.csv')
for line in lines:
    if line[1] == 'male':
        male += 1
    if line[1] == 'female':
        female += 1
        
#print(male)
#print(female)

# Data to plot
labels = 'Male', 'Female'
sizes = [male, female]
colors = ['lightblue', 'pink']
explode = (0.1, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)

plt.axis('equal')
plt.show()

industry = {}
for line in lines:
    if line[3] not in industry:
        industry[line[3]] = 1
    if line[3] in industry:
        industry[line[3]] += 1

industry = {k: v for k, v in sorted(industry.items(), key=lambda item: item[1], reverse=True)}
#industry = {k: industry[k] for k in list(industry)[:10]}

print(industry)
        
plt.figure(figsize=(10, 3))  # width:20, height:3
plt.bar(range(len(industry)), industry.values(), align='edge', width=0.8)
plt.xticks(range(len(industry)), list(industry.keys()), rotation=90)

plt.show()

```
